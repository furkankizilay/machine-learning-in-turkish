### K-Means 

* Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.
* Kümeleme yöntemlerinin amacı benzerlik matrislerini kullanarak, gözlemeri ya da değişkenleri kümelemeye çalışmaktır. Oluşturulmaya çalışan kümelerin kendi içinde homojen, birbirlerine göre heterojen olması beklenir.

- 1) Küme sayısı belirlenir.
- 2) Rastgele k merkez belirlenir.
- 3) Her gözlem için k merkezlere uzaklıklar hesaplanır.
- 4) Her gözlem en yakın olduğu merkeze yani kmeye atanır.
- 5) Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamları yapılır.
- 6) Bu işlem bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamının (total within-cluser variation) minimum olduğu durumdaki gözlemerin kümeleme yapısı nihai kümeleme olarak seçilir.

* Küme içi benzerlik maksimum olmalıdır.
* Kümeler arası benzemezlik te maksimum olmalıdır.

#### Optimum Küme Sayısının Belirlenmesi 

##### Elbow Yöntemi 

```python
ssd = []

K = range(1,30)

for k in K :
    kmeans = KMeans(n_clusters = k).fit(df)
    ssd.append(kmeans.inertia_)
```

```python
plt.plot(K, ssd, "bx-")
plt.xlabel("Farklı K Değerlerine Karşılık Uzaklık Artık Toplamları")
plt.title("Optimum Küme Sayısı İçin Elbow Yöntemi")
```

### Hierarchical Cluster Analysis 

* Amaç gözlemeri birbirlerine olan benzerliklerine göre alt kümelere ayırmaktır.
* Gözlemler dahafazla sayıda alt kümeye ayrılmak isteniyorsa bu durumda hiyerarik kümeleme yöntemleri kullanılır.
* Hiyerarşik kümelemede hem belirli bir sayıda kümeye ayırabiliriz hem de oluşan yeni kümelerin altında da yeni kümelere ayırabiliriz.

* Agglomerative 


* Divisive 


```python
from scipy.cluster.hierarchy import linkage
```

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
# linkage='ward'
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize = (15,10))
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete,
          leaf_font_size = 10);
```

```python
plt.figure(figsize = (10,5))
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete,
           truncate_mode = "lastp",
           p = 10,
           show_contracted = True,
          leaf_font_size = 10);
```

```python
plt.figure(figsize = (15,10))
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_average,
          leaf_font_size = 10);
```

### Principal Component Analysis 

* Temele fikir, çok değişkenli verinin ana özelliklerini daha az sayıda değişken/bileşen ile temsil etmektir.
* Diğer bir ifade ile ; küçük miktarda bir bilgi kaybını göze alıp değişken boyutunu azaltmaktır.

```python
df = pd.read_csv("Hitters.csv")
df.dropna(inplace=True)
df = df._get_numeric_data()
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df = StandardScaler().fit_transform(df)
```

```python
df[0:5,0:5]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca_fit = pca.fit_transform(df)
```

```python
bilesen_df = pd.DataFrame(data = pca_fit, columns = ["birinci_bilesen","ikinci_bilesen"])
```

```python
bilesen_df
```

```python
pca.explained_variance_ratio_ # birinci bilesence veri setinin içerisinde bulunan değişkenliğin yüzde 45'i açıklanabilmiş
# ikinci bileşence %24'ü açıklanabilmiş
```

```python
pca.components_
```

```python
# optimum bileşen sayısı
pca = PCA().fit(df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Bileşen Sayısı")
plt.ylabel("Kümülatif Varyans Oranı");
```

### Model ve Tahmin 

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, r2_score,roc_auc_score, roc_curve,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
import warnings
warnings.filterwarnings("ignore", category = DeprecationWarning)
warnings.filterwarnings("ignore", category = FutureWarning)
```

```python
df = pd.read_csv("diabetes.csv")
y = df["Outcome"]
X = df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
knn_model = KNeighborsClassifier().fit(X_train, y_train)
```

```python
knn_model
```

```python
y_pred = knn_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred) # modelimiz %68 doğru tahmn yapmıştır (ilkel test hatası)
```

```python
print(classification_report(y_test, y_pred))
```

### Model Tuning 

```python
knn = KNeighborsClassifier()
```

```python
knn_params = {"n_neighbors" : np.arange(1,50)}
```

```python
knn_cv_model = GridSearchCV(knn, knn_params, cv = 10).fit(X_train, y_train)
```

```python
knn_cv_model.best_params_
```

```python
knn_cv_model.best_score_
```

```python
# final modeli
knn_tuned = KNeighborsClassifier(n_neighbors = 11).fit(X_train, y_train)
```

```python
y_pred = knn_tuned.predict(X_test)
```

```python
accuracy_score(y_test, y_pred) # başarı % 73'e çıktı
```

```python
knn_tuned.score(X_test, y_test) # y_pred işlemi yapılmadan direkt başarı oranı bulunabilir
```

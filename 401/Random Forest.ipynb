### Model ve Tahmin 

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, r2_score,roc_auc_score, roc_curve,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
import warnings
warnings.filterwarnings("ignore", category = DeprecationWarning)
warnings.filterwarnings("ignore", category = FutureWarning)
```

```python
df = pd.read_csv("diabetes.csv")
y = df["Outcome"]
X = df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
rf_model = RandomForestClassifier().fit(X_train, y_train)
```

```python
# kaç ağaç olduğu
# bölünmelerde göz önüne alınacak değişken sayısı
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### Model Tuning 

```python
X_train.shape
```

```python
rf = RandomForestClassifier()
```

```python
# n_estimators kullanılacak olan ağaç sayısı
# max_features bölünmelerde göz önüne alınacak değişken sayısı
rf_params = {"n_estimators" : [100,200,500,1000],
            "max_features" : [3,5,7,8],
            "min_samples_split" : [2,5,10,20]}
```

```python
rf_cv_model = GridSearchCV(rf, rf_params, cv = 10, n_jobs = -1, verbose = 2).fit(X_train, y_train)
```

```python
rf_cv_model.best_params_
```

```python
# final modeli
```

```python
rf_tuned = RandomForestClassifier(max_features = 7, min_samples_split = 10, n_estimators = 200).fit(X_train, y_train)
```

```python
y_pred = rf_tuned.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# değişken önem düzeyi
```

```python
feature_imp = pd.Series(rf_tuned.feature_importances_,
                       index = X_train.columns).sort_values(ascending = False)

sns.barplot(x = feature_imp, y= feature_imp.index)
plt.xlabel("Değişken Önem Skorları")
plt.ylabel("Değişkenler")
plt.title("Değişken Önem Düzeyleri")
plt.show()
```

## Temel Kavramlar 

### Bağımlı Değişken & Bağımsız Değişken

- Bağımlı Değişken (Hedef Değişken - Çıktı Değişkeni - Yanıt Değişkeni):
    - Makine öğrenmesi problemlerinde tahmin etmek için hedeflediğimiz ana değişkenimizdir.
    - Araç fiyatı tahmin uygulamasında araç fiyatı bağımlı değikendir.
- Bağımsız Değişken :
    - Araç fiyatı tahmin uygulamasında KM, hasar durumu vb. bağımsız değişkendir.

### Öğrenme Türleri : Gözetimli & Gözetimsiz 

- Gözetimli Öğrenme
- Gözetimsiz Öğrenme
- Pekiştirmeli Öğrenme
- Yarı Denetimli Öğrenme

Bağımlı değişken ve bağımlı değişkeni meydana getiren bağımsız değişkenler bir aradaysa bu duruma gözetimli öğrenme denir.

Gözetimsiz öğrenme bağımlı değişnenin çalışmanın içinde olmadığı türdür.
Amaç, gözlem birimlerini birbirlerine bnzer özelliklerine göre bir araya getirmeektir.

### Problem Türleri : Regresyon & Sınıfladırma 

Bağımlı değişken sayısal/sürekli bir değişken ise bu bir regresyon problemleridir.

Bağımlı değişken kategorik bir değeişken ise bu sınıflandırma problemidir.

### Model Eğitmek : Test - Train 

Elimizdeki orjinal veri seti, eğitim seti ve test seti olarak ikiye ayrılır.
Örneğin elimizdeki 1000 gözlemin 800 tanesini algoritmayı eğitmek için kullanırız geride kalanlar ile de eğitmiş olduğumuz algoritmayı test ederiz.

###  Değişken Seçimi (Variable Selection)

Amacımız en az değişkenle en fazla açıklanabilirliği yakalamaya çalışmak.

### Model Seçimi 

* Birincisi : Oluşabilecek değişken kombinasyonları ile oluşturulan modeller arasında en iyi modelin seçilmesi.
* İkincisi : Kurulan birbirinden farklı modeller arasında model seçimi.

Model Neye Göre Seçilir ?
* Regresyon için açıklanabilirlik oranı ve RMSE (hata kareler ortalamasının karekökü) türevi bir değer.
* Sınıfladırma için doğru sınıfladnırma oranı türevi bir değer.

### Aşırı Öğrenme (Overfitting)

Algoritmanın kendisine verilen veri setinin yapısını çok iyi öğrenip görmemiş olduğu yeni veriler üzerinden tahmin yapmak istendğinde başarısız olmasını ifade etmektedir.

### Deterministik Modeller vs Stokastik Modeller 

- Deterministik Modeller : Değişkenler arasında kesin bir ilişki olduğunu varsayar. (y = 2.5x)
- Stokastik Modeller : Olasılıksal modelleridir. (y = 2.5x + tesadüfi hata)

### Doğrusal ve Doğrusal Olmayan Modeller 

## Model Doğrulama (Model Validation) Yöntemleri  

Model kurmak, bağımlı değişken ile bağımsız değişkenler arasındaki ilişkiyi modellemek demektir.

Model doğrulama, modellerin ürettiği sonuçların doğru değerlendirilmesi çalışmalarıdır.

### Holdout Yönetimi (Sınama) 

Orjinal veri seti, eğitim seti ve test seti olarak ikiye ayrılır. Orjınal veri setinin %80'i eğitim setinde eğitilir %20'si test setinde test edilir. Gözlem sayısı çok az ise holdout yöntemi kullanılabiliyor.

### K-Katlı Çapraz Doğrulama (k fold cross validation) 

Orjinal veri seti, eğitim seti ve test set olarak ikiye ayrılır. Eğitim seti k adet parçaya ayrılır daha sonra belirlenen k adet parçadan biri dışarıda bırakılır ve elde kalan diğer parçalar ile model oluşturulur. Oluşan model dışarıda bırakılan parça ile test edilir daha sonra aynı işlem bütün parçalar için tekrar edilir. İşlem sonucunda elde edilen hataların ortlamaları alındığında bu validation (doğrulama) hatası olur. Daha sonra çalışmanın en başında test seti olarak böldüğümüz kısmı da test ederiz.

### Leave One Out 

K-katlı çapraz doğrulamanın özel bir halidir. Örnek sayısı kadar küme sayılır ve model kurulup dışarıda bırakılan test edlilir. Veri sti büyüdükçe uygulanamsı zordur.

### Bootstrap 

Yeniden örenk oluşturucak şekilde çalışır. Orjinal veri setinin içerisinden, orjinal veri stinin içerisindeki gözlem sayısından daha az sayıda gözlemle boostrap örneklemleri oluşturulur, oluşturulan bootstaplarla model oluşturulur, kurulan modeller bir test seti yaklaşımıyla test edilir sonuçları karşılaştırılarak (train ortalması alınarak - testlerin ortalamsı alınarak) sonuçlar değerledirilir.

## Model Başarı Değerlendirme Yöntemleri 

Modellerin tahmin başarılarının değerlendirlmesi

### Regreyon Modelleri için Başarı Değerlendirme 

MSE = \frac{1}{n}\sum _{i=1}^n\:\left(y\:-\:ý\right)^2
* y = Gerçek değerler
* ý = Tahmin edilen değerler
* n = gözlem sayısı
* MSE = Hata kareler ortalamsı

RMSE = \left(\frac{1}{n}\sum \:\:_{i=1}^n\:\left(y\:-\:ý\right)^2\right)^{\left(\frac{1}{2}\right)}
* Hata kareler otlamasının karekökü

MAE = \frac{1}{n}\sum \:\:\:_{i=1}^n\left|y\:-\:ý\right|

* Ortalama mutlak hata


### Sınıfladnırma Modelleri için Başarı Değerlendirme 




- a = True Pozitif (TP)
- b = False Negatif (FN)
- c = False Pozitif (FP)
- d = True Negatif (TN)

+ Doğruluk = (TP + TN) / Hepsi
+ Hata Oranı = (FN + FP) / Hepsi
+ Kesinlik = TP / (TP + FP)
+ Anma : TP / (TP + FN)

ROC Eğrisi : 

* Eğrinin altında kalan alana AUC (Area Under Curve) adılı verilir. Bu alanın büyük olamsı modelin başarılı olduğu anlamına gelir.
* Ortadaki kesikli çizgiler hiç model kurulmasaydı elde edlilecek başarı demektir.

## Yanlılık-Varyans Değiş Tokuşu (Bias-Variance Tradeoff) 

Modellerin tahmin başarılarının değerlendirlimesi

Modelin başarısını değerlenirirken iki tip hata olur birincisi eğitim hatası ikincisi ise tet hatasıdır.
- Eğitim Hatası : Modeli kurmak için kullanılan veri seti üzerinden elde ettiğimiz hata.
- Test Hatası : Kurulan modeli test etmek için kullandığımız veri seti için oluşan hatadır.

Eğitim hatası test hatasının yanlı ve kötü bir temsilcisidir.

Esneklik verinin fonksiyonel yapısının uygun bir şekilde yorumlanmasıdır.



* En sağdakı eğri esnektir.
* Yanlılık gerçek değerler ile tahmin edilen değerler arasındaki farkı ifade eder.
* Varyans değişkenlik demektir yani esnekliktir.

Esneklik neye göre seçilir (doğru model - duşük yanlılık ve düşük varyans) : Esneklik ortalama test hatasına göre seçilir. Ortalama test hatası yüksekse daha esnek model seçilir. Model esnekleştikçe varyans artar, yanlılık azalır belirli bir noktada artarsa overfitting meydana gelir.

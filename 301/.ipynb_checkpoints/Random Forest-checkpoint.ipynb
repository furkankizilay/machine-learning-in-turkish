## Random Forest 

* Temeli birden çok karar ağacının ürettiği tahminlerin bir araya getirileerek değerlendirilmesine dayanır.
* Topluluk öğrenme yötenleri : Birden fazla algoritmanın ya da birde fazla ağacın bir araya gelerek toplu bir şekilde öğrenmesi ve tahmin etmeye çalışmasdır.
* Bagging : Temeli bootstrap yöntemi ile oluşturulan birden fazla kara ağacının rettiği tahminlerin bir araya getirilierek değerlendirlimesine dayanır.

### Model ve Tahmin 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
df = pd.read_csv("Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
rf_model = RandomForestRegressor(random_state = 42).fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

### Model Tuning 

```python
rf_model = RandomForestRegressor(random_state = 42).fit(X_train, y_train)
```

```python
# max_features : bölğnelerde göz önnde bulundurulacak değişken sayısını ifade eder
# max_leaf_nodes :
# min_samples_split : ancak bu kadar gözlem varsa bölmeye devam et
# min_samples_leaf : leaf nodedaki minimum gözlem sayısını ifade eder
# n_estimators : kullanılacak olan ağaç sayısı
```

```python
rf_params = {"max_depth" : [5,8,10],
            "max_features" : [2,5,10],
            "n_estimators" : [200,500,1000,2000],
            "min_samples_split" : [2,10,80,100]}
```

```python
rf_cv_model = GridSearchCV(rf_model, rf_params, cv = 10, n_jobs = -1, verbose = 2).fit(X_train, y_train)
```

```python
rf_cv_model.best_params_
```

```python
rf_tuned_model = RandomForestRegressor(random_state = 42,max_depth = 8,max_features=2,min_samples_split=2,n_estimators=200 ).fit(X_train, y_train)
```

```python
y_pred = rf_tuned_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

### Değişken Önem Düzeyi 

```python
Importance = pd.DataFrame({"Importance" : rf_tuned_model.feature_importances_*100},
                         index = X_train.columns)

Importance.sort_values(by = "Importance",
                      axis = 0,
                      ascending = True).plot(kind = "barh",
                                            color = "r")

plt.xlabel("Variable Importance")
plt.gca().legend_ = None
```

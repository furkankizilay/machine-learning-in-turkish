## K Nearest Neighbor

* Gözlemlerin birbirine olan benzerlikleri üzerinden tahmin yapılır.
* Sınıflandırma ya da regresyon problemlerinde kullanılabilen algoritmadır
* Parametrik olmayan bir öğrenme türüdür.
* Büyük veri setlerinde performans olarak çokta iyi değildir.

### KNN Hesaplama Basamakları 

* Komşu sayısı (k) belirlenir.
* Bilinmeyen nokta ile diğer tüm noktalar arasındaki uzaklıklar hesaplanır.
* Uzaklıkları sırala ve belirlenen k sayısına göre en yakın olan k gözlemi seç.
* Sınıflandırma ise en sık sınıf, regresyon ise ortalama değeri tahmin değeri olarak ver.

### Model ve Tahmin 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
from warnings import filterwarnings
filterwarnings("ignore") 

# önceki bölümlerde alınan uyarı mesajlarını ignore eder
```

```python
df = pd.read_csv("Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
X_train.head()
```

```python
# Model
```

```python
knn_model = KNeighborsRegressor().fit(X_train, y_train)
```

```python
knn_model
```

```python
knn_model.n_neighbors
```

```python
knn_model.metric
```

```python
dir(knn_model)
```

```python
# Tahmin
```

```python
knn_model.predict(X_test)[0:5]
```

```python
y_pred = knn_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred)) # ilkel test hatası
```

### Model Tuning 

```python
# k^yı belirleme : karekök(eğitim boyutu)/2

RMSE = []

for k in range(10) :
    k = k + 1
    knn_model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    RMSE.append(rmse)
    print("k = ",k, " için RMSE değeri : ", rmse)
```

```python
# GridSearchCV
```

```python
# Kullancak olduğumuz makine öğrenmesi algoritmlarında, belirlemeye çalıştığımız hiperparametrelerin değerlerini belirlemek için kullanılır.
```

```python
knn_params = {"n_neighbors" : np.arange(1,30,1)}
```

```python
knn = KNeighborsRegressor()
```

```python
knn_cv_model = GridSearchCV(knn, knn_params, cv = 10).fit(X_train, y_train)
```

```python
knn_cv_model.best_params_
```

```python
# final modeli
knn_tuned = KNeighborsRegressor(n_neighbors = knn_cv_model.best_params_["n_neighbors"]).fit(X_train, y_train)
```

```python
y_pred = knn_tuned.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred)) # final test hatası
```

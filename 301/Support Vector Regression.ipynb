### Model ve Tahmin 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
from warnings import filterwarnings
filterwarnings("ignore") 

# önceki bölümlerde alınan uyarı mesajlarını ignore eder
```

```python
df = pd.read_csv("Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
svr_model = SVR("linear").fit(X_train, y_train) # kernel rbf de olabilirde doğrusal olmayanlar için
```

```python
svr_model.predict(X_train)[0:5]
```

```python
svr_model.predict(X_test)[0:5]
```

```python
svr_model.intercept_
```

```python
svr_model.coef_
```

```python
# ilkel test hatası
y_pred = svr_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

### Model Tuning

```python
svr_model = SVR("linear")
```

```python
svr_model
```

```python
svr_params = {"C" : [0.1, 0.5, 1, 3]}
```

```python
svr_cv_model = GridSearchCV(svr_model, svr_params, cv = 5).fit(X_train, y_train)
```

```python
svr_cv_model.best_params_
```

```python
# svr_cv_model = GridSearchCV(svr_model, svr_params, cv = 5, verbose = 2, n_jobs = -1).fit(X_train, y_train)

# verbose işlem hakkında bilgi verir
# n_jobs işlemcinin son güç çalışmasını sağlar
```

```python
svr_tuned = SVR("linear", C = 0.5).fit(X_train, y_train)
```

```python
y_pred = svr_tuned.predict(X_test)
```
